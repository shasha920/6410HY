```python
paragraph="""Thank you all so very much. Thank you to the Academy.
               Thank you to all of you in this room. I have to congratulate
               the other incredible nominees this year. The Revenant was
               the product of the tireless efforts of an unbelievable cast
               and crew. First off, to my brother in this endeavor, Mr. Tom
               Hardy. Tom, your talent on screen can only be surpassed by
               your friendship off screen … thank you for creating a t
               ranscendent cinematic experience. Thank you to everybody at
               Fox and New Regency … my entire team. I have to thank
               everyone from the very onset of my career … To my parents;
               none of this would be possible without you. And to my
               friends, I love you dearly; you know who you are. And lastly,
               I just want to say this: Making The Revenant was about
               man's relationship to the natural world. A world that we
               collectively felt in 2015 as the hottest year in recorded
               history. Our production needed to move to the southern
               tip of this planet just to be able to find snow. Climate
               change is real, it is happening right now. It is the most
               urgent threat facing our entire species, and we need to work
               collectively together and stop procrastinating. We need to
               support leaders around the world who do not speak for the
               big polluters, but who speak for all of humanity, for the
               indigenous people of the world, for the billions and
               billions of underprivileged people out there who would be
               most affected by this. For our children’s children, and
               for those people out there whose voices have been drowned
               out by the politics of greed. I thank you all for this
               amazing award tonight. Let us not take this planet for
               granted. I do not take tonight for granted. Thank you so very much."""
```

```python
#Tokenize
dataset=nltk.word_tokenize(paragraph)#dataset is a list
```

```python
#Normalize
for i in range(len(dataset)):
  dataset[i]=dataset[i].lower()#convert to characters to lower case
  dataset[i]=re.sub(r'\W',' ',dataset[i])#substiture no-word characters with space
  dataset[i]=re.sub(r'\s+',' ',dataset[i])#we have bunch of paces, get rid of them
```

```python
len(dataset)
```

```python
dataset
```

```python
#Create the histogram or word-frequency, 1st step in BOW model is you need a dictionary of all words mapped with freq
word2count={}
for data in dataset:
  words=nltk.word_tokenize(data)
  for word in words:
    if word not in word2count.keys():#if word not in dictionary, appeared in one sentences not before
      word2count[word]=1
    else:#if word is already in dictionary, you increase count or add 1
      word2count[word]+=1
```

```python
word2count
```

```python
#Now we have to filter words
```

```python
#heapq library, provides an implementation of the heap queue algorithm, also known as the priority queue algorithm
```

```python
import heapq
```

```python
#Select top n words
frequent_words=heapq.nlargest(100,word2count,key=word2count.get)
```

```python
len(frequent_words)
```

```python
#Finally we are going to build our BOW model
#Use the Most Frequent Words in a Model
X=[] #an empty list to enter frequencies
for data in dataset:
  vector=[]
  for word in frequent_words:
    if word in nltk.word_tokenize(data):#checking if our frequent words appear in tokenize data, give it 1
      vector.append(1)
    else:
      vector.append(0) #if not, the vector value would be 0
      #when we are done for one sentence, we want to append for X
  X.append(vector)

```

```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt #plotting libraray for the Python programming language
#and its numerical mathematics extension NumPy
import nltk
```

```python
#building a corpus of documents
corpus=['The sky is blue and beautiful.',
       'Love this blue and beautiful sky!',
       'The quick brown fox jumps over the lazy dog.',
       "A king's breakfast has sausages, ham, bacon, eggs, toast, beans",
       'I love green eggs, ham, sausages and bacon!',
       'The brown fox is quick and the blue dog is lazy!',
       'The sky is very blue and the sky is very beautiful today',
       'The dog is lazy but the brown fox is quick!']
```

```python
corpus
```

```python
labels=['weather','weather','animals','food','food','animals','weather','animals']
```

```python
corpus=np.array(corpus)
```

```python
corpus_df=pd.DataFrame({'Document':corpus,'Category':labels})
```

```python
corpus_df=corpus_df[['Document','Category']]
corpus_df
```

```python
#Preprocessing Text Corpus
```

```python
wpt=nltk.WordPunctTokenizer()
# wordpunct_tokenize will split pretty much at all special symbols and treat them as separate units.
#word_tokenize on the other hand keeps things like "'m"
#For example:I'm, word_tokenize result "I","'m". wordpunct_tokenize result:"I","'","m"
```

```python
stop_words=nltk.corpus.stopwords.words('english')
```

```python
def normalize_document(doc):
  #Lowercase and remove special characters\whitespace
  doc=re.sub(r'[^a-zA-Z\s]',' ',doc,re.I|re.A)#re.I ignore case sensitive, ASCII-only matching
  doc=doc.lower()
  doc=doc.strip()
  #tokenize document
  tokens=wpt.tokenize(doc)
  #filter stopwords out of document
  filtered_tokens=[token for token in tokens if token not in stop_words]
  #re_create document from filtered tokens
  doc=' '.join(filtered_tokens)
  return doc
```

```python
normalize_corpus=np.vectorize(normalize_document)#Basic preprocessing pipeline is ready, we apply our sample corpus
```

```python
norm_corpus=normalize_corpus(corpus)
norm_corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#get bag of words features in sparse format(mostly zeros)
```

```python
cv=CountVectorizer(min_df=0.0,max_df=1.0)
```

```python
cv_matrix=cv.fit_transform(norm_corpus)
cv_matrix
```

```python
#view non_zero feature positions in the sparce matrix
```

```python
print(cv_matrix)
```

```python
#view dense representation, might give error if too big
```

```python
cv_matrix=cv_matrix.toarray()
cv_matrix
```

```python
#get all unique words in the corpus
vocab=cv.get_feature_names_out()
#show document feature matrix
pd.DataFrame(cv_matrix,columns=vocab)
```

```python
#you can set the n-gram range to 1,2 to get unique unigram as well as bigrams
bv=CountVectorizer(ngram_range=(1,2))
bv_matrix=bv.fit_transform(norm_corpus)#Fit to data, then transform it.
bv_matrix=bv_matrix.toarray()
vocab=bv.get_feature_names_out()
pd.DataFrame(bv_matrix,columns=vocab)
```

```python
#you can set the n-gram range to 2,2 to get unique unique unigram as well as bigrams
bv=CountVectorizer(ngram_range=(2,2))
bv_matrix=bv.fit_transform(norm_corpus)
bv_matrix=bv_matrix.toarray()
vocab=bv.get_feature_names_out()
pd.DataFrame(bv_matrix,columns=vocab)
```
