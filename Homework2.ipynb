```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import nltk
```

```python
nltk.download('stopwords')
```

```python
#Build the Text Corpus
corpus_df=pd.read_csv('homework2.csv')
corpus_df
```

```python
corpus=np.array(corpus_df['text'])
corpus
```

```python
#Remove Stop Words and do any cleaning needed
wpt=nltk.WordPunctTokenizer()
stop_words=nltk.corpus.stopwords.words('english')
```

```python
#Create normalized corpus
def normalize_text(text):
  text=re.sub(r'[^a-zA-Z\s]+','',text,re.I|re.A)
  text=text.lower()
  text=text.strip()
  tokens=wpt.tokenize(text)
  filtered_tokens=[token for token in tokens if token not in stop_words]
  text=' '.join(filtered_tokens)
  return text
```

```python
vectorize_corpus=np.vectorize(normalize_text)
vectorize_corpus
```

```python
normalized_corpus=vectorize_corpus(corpus)
normalized_corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tv=TfidfVectorizer(min_df=0.2,max_df=0.8,norm='l2',
                   use_idf=True,smooth_idf=True)
tv_matrix=tv.fit_transform(normalized_corpus)
tv_matrix=tv_matrix.toarray()
vocab=tv.get_feature_names_out()
pd.DataFrame(np.round(tv_matrix,2),columns=vocab)
```
