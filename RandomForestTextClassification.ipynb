```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn import datasets
```

```python
corpus=['The apple is on sale', 'The oranges are on sale','The apple is present','The orange is present']
```

```python
Y=np.array([1,2,1,2])
```

```python
vectorizer=CountVectorizer(min_df=1)
X=vectorizer.fit_transform(corpus).toarray()
```

```python
clf=RandomForestClassifier()
clf.fit(X,Y)
clf.predict(vectorizer.transform(['apple is green']).toarray())
```

```python
from sklearn.datasets import load_wine
#https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_wine.html
wine=load_wine()
X=wine.data
y=wine.target
```

```python
wine.data
```

```python
wine.target
```

```python
wine.feature_names
```

```python
wine.target_names
```

```python
#Build the Ramdom Forest Classifier Model
```

```python
rf=RandomForestClassifier(n_estimators=100,
                          max_depth=3,
                          max_features='sqrt',
                          min_samples_leaf=2,
                          bootstrap=True,
                          n_jobs=-1,
                          random_state=0)
rf.fit(X,y)
```

```python
#In a jupter environment, please rerun this cell to show the HTML representation or trust the notebook.
#On GitHub, the HTML representation is unable to render, please try loading this page with vbviewer.org
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

fig=plt.figure(figsize=(15,10))
plot_tree(rf.estimators_[0],
         feature_names=wine.feature_names,
         class_names=wine.target_names,
         filled=True, impurity=True,
         rounded=True)
```
